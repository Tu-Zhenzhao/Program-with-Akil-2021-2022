# Import Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('original-output.jpg')
imgplot = plt.imshow(img)
#plt.axis("off")
plt.show()
```

```python
img.shape
```

```python
b, g, r= img[:, :, 0], img[:, :, 1], img[:, :, 2]
```

```python
pca = PCA(20)

red_transformed = pca.fit_transform(r)

red_inverted = pca.inverse_transform(red_transformed)

green_transformed = pca.fit_transform(g)

green_inverted = pca.inverse_transform(green_transformed)

blue_transformed = pca.fit_transform(b)

blue_inverted = pca.inverse_transform(blue_transformed)
```

```python
img_compressed = (np.dstack((blue_inverted, green_inverted, red_inverted))).astype(np.uint8)
```

```python
imgplot_2 = plt.imshow(img_compressed)
plt.show()
```
