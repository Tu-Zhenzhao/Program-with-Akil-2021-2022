# Import Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('original-output.jpg')
imgplot = plt.imshow(img)
#plt.axis("off")
plt.show()
```

```python
img.shape
```

```python
b, g, r= img[:, :, 0], img[:, :, 1], img[:, :, 2]
```

```python
pca = PCA(20)

red_transformed = pca.fit_transform(r)

red_inverted = pca.inverse_transform(red_transformed)

green_transformed = pca.fit_transform(g)

green_inverted = pca.inverse_transform(green_transformed)

blue_transformed = pca.fit_transform(b)

blue_inverted = pca.inverse_transform(blue_transformed)
```

```python
img_compressed = (np.dstack((red_inverted, red_inverted, red_inverted))).astype(np.uint8)
```

```python
imgplot_2 = plt.imshow(img_compressed)
plt.show()
```

```python
df_b=b/255
df_g=g/255
df_r=r/255
```

```python

df_red_transformed = pca.fit_transform(df_r)

df_red_inverted = pca.inverse_transform(red_transformed)

df_green_transformed = pca.fit_transform(df_g)

df_green_inverted = pca.inverse_transform(green_transformed)

df_blue_transformed = pca.fit_transform(df_b)

df_blue_inverted = pca.inverse_transform(blue_transformed)
```

```python
df_img_compressed = (np.dstack((df_red_inverted, df_red_inverted, df_red_inverted))).astype(np.uint8)
```

```python
imgplot_3 = plt.imshow(df_img_compressed)
plt.show()
```
